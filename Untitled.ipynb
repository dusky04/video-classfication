```python
import torch
from torchvision import transforms
```

```python
# In dataset.py - modify train_transform
transform = transforms.Compose([
    transforms.ConvertImageDtype(torch.float32),
    transforms.RandomApply([transforms.RandomRotation(10)], p=0.3),
    transforms.RandomApply([transforms.RandomAffine(0, translate=(0.1, 0.1))], p=0.3),
    transforms.RandomApply([transforms.ColorJitter(0.2, 0.2, 0.2, 0.1)], p=0.8),
    transforms.RandomGrayscale(p=0.2),
    transforms.RandomApply([transforms.GaussianBlur(kernel_size=(5, 5))], p=0.5),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
b = transform(ds.load_video_frames(0))
```
